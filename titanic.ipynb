### Data Preprocessing

```python
print('train_df.shape :', train_df.shape)
print('test_df.shape :', test_df.shape)
```

```python
input_cols = list(train_df.columns.drop(["Name","Ticket","Cabin","Survived"]))
target_col = 'Survived'
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_col].copy()

test_inputs = test_df[input_cols].copy()
```

#### Encoding

```python
#Binary Encoding --> Sex
sex_code = {"male":1,"female":0}
train_inputs["sex_code"] = train_inputs["Sex"].map(sex_code)
test_inputs["sex_code"]=test_inputs["Sex"].map(sex_code)
```

```python
train_inputs.head()
```

```python
corr = train_inputs.corr(numeric_only=True)
corr
```

```python
sns.heatmap(data=corr,cmap="Blues",fmt = ".2f",annot=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder= OneHotEncoder(sparse_output=False,handle_unknown='ignore')
```

```python
encoder.fit(train_inputs[["Embarked"]])
```

```python
encoder.categories_
```

```python
encoded_cols = list(encoder.get_feature_names_out(["Embarked"]))
```

```python
encoded_cols
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[["Embarked"]])
test_inputs[encoded_cols] = encoder.transform(test_inputs[["Embarked"]])
```

```python
encoded_cols.append("sex_code")
```

```python
train_inputs
```

#### Imputing

```python
from sklearn.impute import SimpleImputer
```

```python
train_inputs[["Age","Fare"]].isnull().sum()
```

```python
imputer = SimpleImputer(strategy="mean")
```

```python
imputer.fit(train_inputs[["Age","Fare"]])
```

```python
train_inputs[["Age","Fare"]] = imputer.transform(train_df[["Age","Fare"]])
test_inputs[["Age","Fare"]] = imputer.transform(test_df[["Age","Fare"]])
```

```python
train_inputs[["Age","Fare"]].isnull().sum()
```

```python
train_df.describe()
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

* First of all need to get numeric columns. But sex_code column is already binary encoded so we don't need to scale it.

```python
numeric_cols = list(train_inputs.select_dtypes(include=["float64", "int64"]).columns.drop(encoded_cols))
```

```python
numeric_cols
```

```python
scaler.fit(train_inputs[numeric_cols])
```

* Fit the model only on the training data to prevent data leakage.

```python
train_inputs[numeric_cols] =scaler.transform(train_inputs[numeric_cols])
test_inputs[numeric_cols] =scaler.transform(test_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols].describe().round(3) 
```

### Logistic Regression

```python
X_train = train_inputs[numeric_cols+encoded_cols]
X_test = test_inputs[numeric_cols+encoded_cols]

Y_train = train_targets
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import  cross_validate
```

```python
model1 = LogisticRegression(solver='lbfgs')
```

```python
cross_val_model1 = cross_validate(model1,X_train,Y_train,cv=5,return_train_score=True)
print("Test Accuracy Scores:", cross_val_model1['test_score'])
print("Train Accuracy Scores:", cross_val_model1['train_score'])
print("Average Train Accuracy:", cross_val_model1['train_score'].mean())
print("Average Test Accuracy:", cross_val_model1['test_score'].mean())
```

```python
model1.fit(X_train,Y_train)
```

```python
predictions = model1.predict(X_train)
score = accuracy_score(predictions,Y_train)
```

```python
print(f"Accuracy of model on train data set: {score}")
```

```python
test_preds_df = pd.DataFrame({"PassengerId":list(test_df["PassengerId"]),"Survived":model1.predict(X_test)})
test_preds_df.to_csv('test_predicts_1.csv',index=False)
```

```python
print("Accuracy Score of Test set from Kaggle: 0.77033")
```

#### Hyperparameter Tuning with Grid Search Cross Validation

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.1, 1, 10],
    'solver': ['liblinear', 'saga'],
    'max_iter': [1000,2000],
    'penalty': ['l1', 'l2'],
    'class_weight': [None, 'balanced']
}
```

```python
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=5, n_jobs=-1)
```

```python
grid_search.fit(X_train,Y_train)
```

```python
model2= LogisticRegression(**grid_search.best_params_)
grid_search.best_params_
```

```python
cross_val_model2 = cross_validate(model2,X_train,Y_train,cv=5,return_train_score=True)
print("Test Accuracy Scores:", cross_val_model2['test_score'])
print("Train Accuracy Scores:", cross_val_model2['train_score'])
print("Average Train Accuracy:", cross_val_model2['train_score'].mean())
print("Average Test Accuracy:", cross_val_model2['test_score'].mean())
```

```python
model2.fit(X_train,Y_train)
```

```python
test_preds_df2 = pd.DataFrame({"PassengerId":list(test_df["PassengerId"]),"Survived":model2.predict(X_test)})
```
